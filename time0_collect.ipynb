```python
import os
import shutil

# EDIT HERE
source_root = ".data/multipleview/painter"
destination_folder = "time0_painter"

if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# EDIT DEPENDEING ON FOLDER/FILENAME
for i in range(1, 17):
    cam_folder = f"cam{i:02d}"
    source_file = os.path.join(source_root, cam_folder, "frame_00001.jpg")
    if os.path.exists(source_file):
        shutil.copy(source_file, os.path.join(destination_folder, f"render_time0_{i}.png"))
    else:
        print(f"File {source_file} does not exist.")

print("Images have been collected.")
```
